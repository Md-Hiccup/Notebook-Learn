```python
import pandas as pd

######################### 
# create DF using csv file
######################### 

df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/3_different_ways_of_creating_dataframe/weather_data.csv')
df
```

```python
######################### 
# create DF using python dictionary
######################### 
weather_data = {
    'day': ['1/1/2020', '1/2/2020', '1/3/2020', '1/4/2020'],
    'event': ['Rain', 'Sunny', 'Snow', 'Snow'],
    'windspeed': [6,7,2,7],
    'temperature': [32,35,28,24]
}

df = pd.DataFrame(weather_data)
df
```

```python
######################### 
# create DF using list of tuples
#########################
weather_data = [
    ('1/1/2020', 32, 6, 'Rain'),
    ('1/2/2020', 35, 7, 'Sunny'),
    ('1/3/2020', 28, 2, 'Snow'),
]
df = pd.DataFrame(weather_data, columns=['day', 'temperature', 'windspeed', 'event'])
df
```

```python
######################### 
# create DF using list of dictionary
######################### 
weather_data = [
    {'day': '1/1/2020', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/1/2020', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/1/2020', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]

df = pd.DataFrame(weather_data)
df
```
