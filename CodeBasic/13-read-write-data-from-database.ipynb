```python
"""
Read database records into pandas dataframe and write it back

1. pd.read_sql_table
2. pd.read_sql_query
3. df.to_sql
4. pd.read_sql(which is just a wrapper for first two methods)
"""
```

```python
import pandas as pd
import sqlalchemy
```

```python
# sqlalchemy.create_engine('mysql+pymysql://<user>:@<host>:<port>/<db name>')
engine = sqlalchemy.create_engine('mysql+pymysql://root:root@localhost:3306/pandas')
```

```python
#####################
# To read the sql table
# read_sql_table()

df = pd.read_sql_table('customers', engine, columns=['name'])
df
```

```python
#####################
# To run the query 
# read_sql_query()

query = '''
SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM pandas.customers INNER JOIN pandas.orders
 ON customers.id=orders.customer_id
'''
df = pd.read_sql_query(query, engine)
df
```

```python
#######
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/21_sql/customers.csv')
df
```

```python
# To rename the columns of dataframe
df.rename(columns={
    'Customer Name': 'name',
    'Customer Phone': 'phone_number'
}, inplace=True)
df
```

```python
#####################
# To write into sql
# to_sql()

df.to_sql(
    'customers', 
    con=engine,
    index=False,
    if_exists='append'
)
```

```python
#####################
# read_sql()  provides convenience for read_sql_table() and read_sql_query()

df = pd.read_sql('customers', engine)
df
```

```python
query = '''
SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM pandas.customers INNER JOIN pandas.orders
 ON customers.id=orders.customer_id
'''
df = pd.read_sql(query, engine)
df
```
